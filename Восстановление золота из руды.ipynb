### Расчёт эффективности

```python
c = gold_train['rougher.output.concentrate_au']
f = gold_train['rougher.input.feed_au']
t = gold_train['rougher.output.tail_au']
```

```python
recovery = ((c * (f - t)) / (f * (c - t))) * 100 
```

```python
mae = mean_absolute_error(gold_train['rougher.output.recovery'], recovery)
```

### Суммарные концентрации

```python
def sum_concetrate(stage, gold):
    sum_stage = 0
    for name in stage: 
        sum_stage += gold[name] 
    return sum_stage
```

```python
rougher_input = ['rougher.input.feed_ag', 'rougher.input.feed_pb', 'rougher.input.feed_sol', 'rougher.input.feed_au']
sum_rougher_input = sum_concetrate(rougher_input, gold_full)
```

```python
rougher_output = ['rougher.output.concentrate_ag', 'rougher.output.concentrate_pb', 'rougher.output.concentrate_sol', 'rougher.output.concentrate_au']
sum_rougher_output = sum_concetrate(rougher_output, gold_full)
```

```python
primary_cleaner = ['primary_cleaner.output.concentrate_ag', 'primary_cleaner.output.concentrate_pb', 'primary_cleaner.output.concentrate_sol', 'primary_cleaner.output.concentrate_au']
sum_primary_cleaner = sum_concetrate(primary_cleaner, gold_full)
```

```python
final = ['final.output.concentrate_ag', 'final.output.concentrate_pb', 'final.output.concentrate_sol', 'final.output.concentrate_au']
sum_final = sum_concetrate(final, gold_full)
```

```python
plt.figure(figsize=[25, 25])

plt.subplot(4, 1, 1)
plt.title(f'График суммарных концентраций на этапах ROUGHER, PRIMARY_CLEANER, FINAL', fontsize=13)
plt.xlabel('Суммарная концентрация', fontsize=11)
plt.ylabel('Количество наблюдений', fontsize=11)
sns.histplot(sum_rougher_input, color='black', label='Сырье', kde=False)
sns.histplot(sum_rougher_output, color='g', label='Концентрат после флотации', kde=False)
sns.histplot(sum_primary_cleaner, color='y', label='Концентрат после первой очистки', kde=False)
sns.histplot(sum_final, color='r', label='Финальный концентрат', kde=False)
    
plt.grid(True)

plt.legend(title='Этап', loc=2, fontsize=8)
plt.show();
```

**Вывод:**

- Суммарная концентрация веществ увеличивается к финальному этапу, а диапазон распределения суммарной концентрации веществ уменьшается.
- Есть большое количество выбросов  в нулевом значение.

### Функция для вычисления итогового sMAPE

```python
def smape(actual, prediction):
    return 1 / len (actual) * np.sum (np.abs (prediction-actual) / ((np.abs (actual) + np.abs (prediction)) /2)) * 100
```

```python
smape_scor = make_scorer(smape)
```

```python
def final_smape(rougher, final):
    return 0.25 * rougher + final * 0.75
```

### Обучение моделей

#### Обучение модели DecisionTreeRegressor

```python

def regressor_tree_train(features, target):
    best_score = 100
    best_depth = None
    for depth in range(1, 13):
        model = DecisionTreeRegressor(random_state=12345, max_depth=depth)
        temp_score = cross_val_score(model, features, target, cv=5, scoring=smape_scor)
        if temp_score.mean() < best_score:
            best_score = temp_score.mean()
            best_depth = depth
            
            
    return best_score, best_depth
```
